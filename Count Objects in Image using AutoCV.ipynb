# Count Objects in Image using AutoCV

In that project I will try to count objects in an image with Auto Copmuter Vision from cvlib library. 
CVlib is an open-source computer vision library designed to provide easy-to-use and high-level APIs for computer vision tasks like object detection, face detection, and more. It simplifies complex operations in computer vision, allowing developers to build applications without diving deep into intricate details.
 Let's start.

```python
#Importing necessary libraries
import cv2
import matplotlib.pyplot as plt
import cvlib as cv
from cvlib.object_detection import draw_bbox
```

```python
#Reading a cat image
img = cv2.imread('cats.jpeg')
```

```python
#Displaying the image
plt.imshow(img)
```

```python
#Detecting objects
box,label,count=cv.detect_common_objects(img)
```

```python
#Displaying the detected objects with a bounding box
output=draw_bbox(img,box,label,count)
```

```python
#Displaying the output
plt.figure(figsize=(10,10))
plt.imshow(output)
```

```python
#Count of cats
c_count=label.count('cat')
print(f"Counts of cats in the image is :{c_count} ")
```
